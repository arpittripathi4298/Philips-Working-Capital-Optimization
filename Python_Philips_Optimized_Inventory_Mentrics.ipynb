```python
import numpy as np
import pandas as pd
```

```python
Philips_Financial_Cleaned_Data=pd.read_csv('C:/Users/arpit/OneDrive/Desktop/Philips/Fact_Inventory_Raw_Results.csv')
Philips_Financial_Cleaned_Data
```

#### Define a standard devation proxy for suppier lead time based on historical variations.
#### (Assuming a baseline historical standard deviation of 15% of the expected lead time)

```python
Philips_Financial_Cleaned_Data['Lead_Time_Std_Dev']=Philips_Financial_Cleaned_Data['Expected_Lead_Time_Days']*0.15
Philips_Financial_Cleaned_Data
```

#### Calculate Optimal Safety Stock using the standard formula
#### Z = 1.96 represents a 95% service level requirement for Philips healthcare/consumer products

```python
z=1.96
Philips_Financial_Cleaned_Data['Optimal_Safty_Stock']=np.sqrt((Philips_Financial_Cleaned_Data['Expected_Lead_Time_Days']*(Philips_Financial_Cleaned_Data['Demand_Std_Dev']**2))+((Philips_Financial_Cleaned_Data['Average_Daily_Demand']**2)*(Philips_Financial_Cleaned_Data['Lead_Time_Std_Dev']**2)))*z
Philips_Financial_Cleaned_Data
```

```python
Philips_Financial_Cleaned_Data['Optimal_Safety_Stock']=Philips_Financial_Cleaned_Data['Optimal_Safty_Stock'].round(0).astype(int)
Philips_Financial_Cleaned_Data
```

#### Calculate Current Value and Actual Carrying Cost (WACC + Storage = 22%)

```python
Philips_Financial_Cleaned_Data['Current_Inventory_Value_EUR']=Philips_Financial_Cleaned_Data['On_Hand_Inventory']*Philips_Financial_Cleaned_Data['Unit_Cost_EUR']
Philips_Financial_Cleaned_Data['Annual_Carrying_COst_EUR']=Philips_Financial_Cleaned_Data['Current_Inventory_Value_EUR']*0.22
Philips_Financial_Cleaned_Data
```

#### Flag SKUs that are overstocked vs understocked

```python
Philips_Financial_Cleaned_Data['Inventory_Status']=np.where(Philips_Financial_Cleaned_Data['On_Hand_Inventory']<Philips_Financial_Cleaned_Data['Optimal_Safety_Stock'],'Understocked / Risk',np.where(Philips_Financial_Cleaned_Data['On_Hand_Inventory']>(Philips_Financial_Cleaned_Data['Optimal_Safety_Stock']*2),'Overstocked / Excess Capital','Optimal'))
Philips_Financial_Cleaned_Data
```
